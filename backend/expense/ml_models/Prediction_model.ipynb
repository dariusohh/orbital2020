## **Data Cleaning**

```python
for objects in sample_data:
    del objects["id"]
    del objects["username"]
    del objects["name"]
    objects["amount"] = float(objects["amount"])
```

```python
data=  pd.DataFrame(sample_data)
```

```python
data["created_at"] = pd.to_datetime(data["created_at"]) + timedelta(hours=8)
data["created_at"] = data["created_at"].dt.tz_localize(None)
```

```python
data = data.rename(columns={'created_at': 'ds', 'amount': 'y'})
```

```python
data.head()
```

## **Expense/Revenue Forecasting**

```python
expense = data[data["y"] < 0] #Change to > 0 for Revenue Forecast
```

```python
gm_prophet = fbprophet.Prophet(changepoint_prior_scale=0.15, 
                               daily_seasonality=True,weekly_seasonality=True,
                               yearly_seasonality=True,holidays=public_holidays)
gm_prophet.fit(expense)
```

```python
gm_forecast1 = gm_prophet.make_future_dataframe(periods=7, freq='M')
```

```python
gm_forecast = gm_prophet.predict(gm_forecast1)
```

```python
gm_prophet.plot(gm_forecast, xlabel = 'Date', ylabel = 'Amount')
```

```python
gm_forecast
```

```python
# Plot the trends and patterns
gm_prophet.plot_components(gm_forecast)
```

## **Model Evaluation through Cross Validation**

```python
d=cross_validation( model = gm_prophet, initial='20 days', period='8 days', horizon = '5 days')
```

```python
d.head()
```

```python

df_p = performance_metrics(d)
df_p.head()
```

```python

fig = plot_cross_validation_metric(d, metric='mdape')
# downward trend means as more data inout, model becomes more accurate
print( d)
performance_metrics_results =performance_metrics(d)
print (performance_metrics_results)
```

```python
#Notes for Mape:
#Dots show the absolute percent error for each prediction in df_cv. 
#The blue line shows the MAPE, where the mean is taken over a rolling window of the dots.

#plotting the Mdape graph :
arr_mse =[]
for index, row in performance_metrics_results.iterrows():
        arr_mse.append({"days":row["horizon"],"value":row["mdape"]})
       
```
